```python
def cost(y_train,theta,X_train):
  z = dot(X_train,theta)
  cost0 = y_train.T.dot(log(sigmoid(z)))
  cost1 = (1-y_train).T.dot(log(1-sigmoid(z)))
  cost = -((cost1 + cost0))/len(y_train) 
  return cost
```

```python
def gradient_descent(X_train, y_train, theta, alpha, iterations):
    theta_tr=np.transpose(theta)
    cost_history = np.zeros(iterations)
    
    for i in range(iterations):
        # print('theta= ',theta)
        g = X_train.dot(theta_tr)
        hx=sigmoid(g)
        # print('predictions= ',predictions)
        errors = np.subtract(hx,y_train)
        # print('errors =',errors)
        sum_delta = (alpha/m)*X_train.transpose().dot(errors)
        # print('sum_delta =',sum_delta)
        theta = np.subtract(theta,sum_delta)
        
        cost_history[i] = cost(y_train, theta,X_train )
    return theta, cost_history
```
